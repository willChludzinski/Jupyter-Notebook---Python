# pandas

## Why pandas
- open source, free, comes along with anaconda distribution
- makes importing and analyzing data much easier
- good I/O capabilites like reading data fom a Database
- builds on packages like; numpy, matplotlib
- gives a single, convinent package to do most of your data analysis and visualization
- to work with tabular data, pandas has data structures; series, dataframe
- pandas has series, dataframe 

## pandas dataframe
* numpy is great for matrix algebra, but pandas dataframe built on numpy makes it easier to import and analyze tabular data
* pandas' dataframe is an R-style dataframe with column names
* dataframe has labels for columns (vs just index in numpy)
* dataframe can contain different data types at different columns (remember numpy contains a single data type)

```python
# another example of preparing data (and everything before)
numbers_numeric = [1, 2, 3, 4, 5]
numbers_english = ['one', 'two', 'three', 'four', 'five']
numbers_spanish = ['uno', 'dos', 'tres', 'cuatro', 'cinco']
numbers_turkish = ('bir', 'iki', 'uc', 'dort', 'bes')

resultZip = zip(numbers_numeric, numbers_english, numbers_spanish, numbers_turkish)
print(resultZip,'\n')

set4resultZip = set(resultZip)
print(set4resultZip,'\n')

resultZip = zip(numbers_numeric, numbers_english, numbers_spanish, numbers_turkish)
list4resultZip = list(resultZip)
print(list4resultZip,'\n')

# 3) convert data set to pandas dataframe
df_numbers = pd.DataFrame(data=list4resultZip, columns=['numbers_numeric', 'numbers_english', 'numbers_spanish', 'numbers_turkish'])
print('df_numbers.shape:', df_numbers.shape) # (5,4) - 5 rows, 4 columns
df_numbers
```
