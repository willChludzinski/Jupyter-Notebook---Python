```python
import matplotlib.pyplot as plt
import numpy as np
t=np.linspace(0,2*np.pi, 50)

x=np.sin(t)
y=np.cos(t)

plt.plot(x)
plt.title("sine plot")

plt.figure()
plt.plot(y)
plt.title("cosine plot")
```

```python
x=np.linspace(0, 2*np.pi, 50)
y1=np.sin(x)
y2=np.sin(2*x)

plt.figure()   
plt.plot(y1)
plt.plot(x,y1)
```
